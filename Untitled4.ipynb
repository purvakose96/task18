```python
from sklearn.preprocessing import StandardScaler

# Step 1: Load Data
filepath = "/content/ford_car_dataset.csv"
df = pd.read_csv(filepath)

# Step 2: Remove Missing Values and Duplicates
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Step 3: Histogram
numeric_cols = ['year', 'price', 'mileage', 'tax', 'mpg', 'engineSize']

for col in numeric_cols:
    plt.figure(figsize=(5,4))
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.title(col)
    plt.show()

# Step 4: Count Plots
categorical_cols = ['model', 'transmission', 'fuelType']

for col in categorical_cols:
    plt.figure(figsize=(8,5))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.show()

# Step 5: Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.show()

# Step 6: Input and Output Features
X = df.drop('price', axis=1)
y = df['price']

# Step 7: One-Hot Encoding
X = pd.get_dummies(X,
                   columns=['model', 'transmission', 'fuelType'],
                   drop_first=True)

# Step 8: Scaling
numeric_features = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

scaler = StandardScaler()
X[numeric_features] = scaler.fit_transform(X[numeric_features])

print("Final Preprocessed Data")
print(X.head())

print("\nTarget Variable")
print(y.head())
```
