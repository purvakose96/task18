```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Step 1: Load Dataset
filepath = "/content/insurance - insurance.csv"
df = pd.read_csv(filepath)

# Step 2: Remove Missing & Duplicate Values
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Step 3: Histogram
numeric_cols = ['age','bmi','children','expenses']

for col in numeric_cols:
    plt.figure(figsize=(5,4))
    plt.hist(df[col], bins=20, edgecolor='black')
    plt.title(col)
    plt.show()

# Step 4: Count Plots
categorical_cols = ['sex','smoker','region']

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col,
                  data=df,
                  order=df[col].value_counts().index)
    plt.title(col)
    plt.xticks(rotation=30)
    plt.show()

# Step 5: Correlation Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df[numeric_cols].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.show()

# Step 6: Input and Output Features
X = df.drop('expenses', axis=1)
y = df['expenses']

# Step 7: One-Hot Encoding
X = pd.get_dummies(X,
                   columns=['sex','smoker','region'],
                   drop_first=True)

# Step 8: Standard Scaling
numeric_features = ['age','bmi','children']

scaler = StandardScaler()
X[numeric_features] = scaler.fit_transform(X[numeric_features])

print("Final Preprocessed Data")
print(X.head())

print("\nTarget Variable")
print(y.head())
```
